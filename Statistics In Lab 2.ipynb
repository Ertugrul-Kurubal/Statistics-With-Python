```python
stats.linregress(tv_hours, GPA) # y = -0,295x + 3,787
```

```python
x_new = np.linspace(0,5,100)
y_new = intercept + slope * x_new
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
ax.plot(x_new, y_new)
ax.set_xlabel("TV Hours")
ax.set_ylabel("GPA")
ax.axis("tight");
```
